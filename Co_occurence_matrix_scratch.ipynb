```
img=cv2.imread('bot.jpg') #  # Pic Credit Photo by Alex Knight on Unsplash
scale_percent = 10 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
# resize image
resized = cv2.resize(img, dim)
print('Previous Dimensions:\n',img.shape)
print('Resized Dimensions:\n ',resized.shape)
print('\n') 
cv2_imshow(resized)
```

```
img_gray=resized[:,:,0]
cv2_imshow(img_gray)
```

```
img_gray_reshaped=img_gray[:300,:400]
cv2_imshow(img_gray_reshaped)
```

```
temp=img_gray_reshaped[:50,:50]
cv2_imshow(temp)
```

```
# pixel values- 0 to 255
C = np.zeros(shape=(50,50))
C.shape
```

```
for i in range(C.shape[0]):
    for j in range(C.shape[1]):
        cnt=0
        print('Iteration:',i,j)
        for x in range(img_gray_reshaped.shape[0]):
            for y in range(img_gray_reshaped.shape[1]):
                if x+1<img_gray_reshaped.shape[0] and y+1<img_gray_reshaped.shape[1]:
                    if img_gray_reshaped[x,y]==i and img_gray_reshaped[x+1,y+1] ==j:
                        cnt=cnt+1 
        C[i,j]=cnt               
```

```
C
```

```
k=sum(C.reshape(2500,1))
```

```
C=C/k
```
