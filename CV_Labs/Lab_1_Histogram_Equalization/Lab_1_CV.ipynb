```
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
from numpy import random
from astropy.visualization import make_lupton_rgb
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.style.use('seaborn-white')
```

# Show Image

```
img=cv2.imread('lenna.png')
```

```
img.shape, img.dtype, type(img)
```

```
cv2_imshow(img)
```

# Image After Resizing

```
scale_percent = 40 # percent of original size
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)
# resize image
resized = cv2.resize(img, dim)
print('Previous Dimensions:\n',img.shape)
print('Resized Dimensions:\n ',resized.shape)
print('\n') 
cv2_imshow(resized)
```

# Displaying All 3 Channels Separately

```
cv2_imshow(resized[:,:,0])
cv2_imshow(resized[:,:,1])
cv2_imshow(resized[:,:,2])
```

# Plotting Histogram for each channel

```
plt.figure(figsize=(10,10))
hist1,bins = np.histogram(resized[:,:,0].flatten(),256,[0,256])
hist2,bins = np.histogram(resized[:,:,1].flatten(),256,[0,256])
hist3,bins = np.histogram(resized[:,:,2].flatten(),256,[0,256])
plt.plot(hist1,'r')
plt.plot(hist2,'g')
plt.plot(hist3,'b')
plt.legend(['Channel_1','Channel_2','Channel_3'],loc='upper left')
```

# Histogram Equalization

```
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('lenna.png')
print('Image Shape:',img.shape,'\n')
print(img.flatten().shape)
cv2_imshow(img)
```

# Channelwise Histogram Equalization

```
temp=img
for c in range(0,3):
    temp[:,:,c]=cv2.equalizeHist(temp[:,:,c])
    print('Channel:',c,'\n')
    cv2_imshow(temp[:,:,c])
```
