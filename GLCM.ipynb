# GLCM Featuers for Image Classification based on texture

```
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
from IPython.display import Image
from skimage.feature import greycomatrix, greycoprops
import matplotlib.pyplot as plt
```

```
sky=cv2.imread('8.PNG',0)
grass=cv2.imread('9.PNG',0)
brick=cv2.imread('10.PNG',0)
print(sky.shape, grass.shape, brick.shape)
```

```
GLCM_sky = greycomatrix(sky, [1], [0, np.pi/2], levels=256 ,normed=True, symmetric=True)
GLCM_grass = greycomatrix(grass, [1], [0, np.pi/2], levels=256 ,normed=True, symmetric=True)
GLCM_brick = greycomatrix(brick, [1], [0, np.pi/2], levels=256 ,normed=True, symmetric=True)
# P4-D ndarray
#The grey-level co-occurrence histogram. The value P[i,j,d,theta] is the number of times that grey-level j 
#occurs at a distance d and at an angle theta from grey-level i. 
print(GLCM_sky.shape,GLCM_grass.shape, GLCM_brick.shape)
```

```
plt.plot(GLCM_sky[:,:,0,0])
plt.title()
plt.figure()
plt.plot(GLCM_grass[:,:,0,0])
plt.figure()
plt.plot(GLCM_brick[:,:,0,0])
```
